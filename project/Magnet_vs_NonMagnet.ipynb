```python
# Feature engineering : Assign magnet and non magnet categories by 
# creating another column named 'is_magnet' which can be yes or no
df_elementary['Magnet_Status']=df_elementary['TYPE'].apply(lambda x : 'Yes' if 'Magnet' in str(x) else 'No')
```

```python

desired_scores = df_elementary['subject'].value_counts().head(10).index.to_list()
print(desired_scores)
```

```python
df_scores = df_elementary[df_elementary['subject'].isin(desired_scores)]
df_scores.head(10)
```

```python

##Lets further remove unwanted column data from our dataset

df_scores =df_scores.drop(columns=['GRADELEVEL','MAGNETPROGRAM', 'TYPE',
       'reporting_year', 'lea_name', 'school_code',])
```

```python
df_scores.columns
```

```python
# Convert all the values in glp_pct into numeric values
df_scores['glp_pct'] = pd.to_numeric(df_scores['glp_pct'], errors='coerce')
```

```python
# Analysing total grade level profeciency of magnet and non magnet schools

df_mag_nonmag = df_scores.pivot_table(index='Magnet_Status',columns='subject',values='glp_pct',aggfunc='mean')
df_mag_nonmag
```

```python
#Spot which subjects have the biggest performance gap between Magnet and Non-Magnet schools.

#See if Magnet schools consistently outperform, or if the difference is subject-specific.
import matplotlib.pyplot as plt

ax=df_mag_nonmag.T.plot(kind='bar', figsize=(10,6))
plt.title('Average Grade-Level Proficiency (GLP%) by Subject: Magnet vs Non-Magnet')
plt.ylabel('GLP %')
plt.xlabel('Subject')
plt.xticks(rotation=45)
plt.ylim(0,100)
plt.legend(title='Magnet Status')

# Add percent value labels on top of bars
for p in ax.patches:
    height = p.get_height()
    ax.text(
        p.get_x() + p.get_width() / 2,  # x position: center of the bar
        height + 0.5,                   # y position: a bit above the bar
        f'{height:.1f}%',               # text to display, formatted to 1 decimal place
        ha='center',                   # horizontal alignment
        va='bottom',                   # vertical alignment
        fontsize=6,
        fontweight='bold'

    )
plt.tight_layout()
plt.show()
```
