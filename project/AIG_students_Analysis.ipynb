```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_schooldata = pd.read_csv('C:\\Users\\aksha\\WCPSS_Data_Analysis\\data\\Wake_County_Public_Schools.csv')
df_schooldata.head(10)
```

```python
df_schooldata = df_schooldata.drop(columns=['X','Y','OBJECTID','TAG','PHONE','ADDRESSNUM','ZIPCODE','MAGNETPROGRAM','WEBSITE','DISTRICT'])
```

```python
ax=sns.barplot(data = df_AIG_area, x='CITY',y='AIG KI-GR13 PCT',hue='AIG KI-GR13 PCT',palette="dark:#5A9_r")
sns.despine()
plt.xticks(rotation=45 ,ha='right')
plt.xlabel('City within wake county')
plt.ylabel('Average % of AIG students in schools')
plt.title('Percentage of students identified as AIG across Cities in Wake County',fontsize=10)

for bar in ax.patches:

    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width()/2,
        height+0.1,
        f'{height:.1f}%',
        ha='center',
        va='bottom',
        fontsize='8',
        fontweight='bold'
    )
    
plt.legend("")
plt.tight_layout()
```

```python
df_relationship = df_relationship.drop(columns=['SCHOOL CODE','NAME','CALENDAR','ADM Month2'])
```

```python
df_relationship['SCHOOL NAME'].value_counts()
```

```python
df_relationship[['glp_pct', 'AIG KI-GR13 PCT']].info()
```

```python
df_corr = df_relationship[['glp_pct', 'AIG KI-GR13 PCT']].dropna()
```

```python
df_corr['glp_pct'] = pd.to_numeric(df_corr['glp_pct'],errors='coerce')

```

```python
df_corr['AIG KI-GR13 PCT']=df_corr['AIG KI-GR13 PCT'].str.strip('%').astype('float')
```

```python
df_corr.head()
```

```python
# Using Pearson relationship to find any relationship between GLP% and number of AIG  students


correlation = df_corr['glp_pct'].corr(df_corr['AIG KI-GR13 PCT'])
print(f"Correlation between GLP% and AIG count: {correlation:.2f}")
```

```python

sns.scatterplot(data=df_corr, x='AIG KI-GR13 PCT', y='glp_pct')
plt.title('GLP% vs AIG Student Count per School')
plt.xlabel('AIG Student percent')
plt.ylabel('GLP%')
plt.grid(True)
plt.show()

sns.regplot(data=df_corr, x='AIG KI-GR13 PCT', y='glp_pct', scatter_kws={'alpha':0.5})
plt.title('Correlation between AIG Count and GLP%')
plt.xlabel('Number of AIG Students')
plt.ylabel('Grade-Level Proficiency (%)')
plt.grid(True)
plt.show()
```

```python
import scipy
correlation1 = df_corr['glp_pct'].corr(df_corr['AIG KI-GR13 PCT'],method='spearman')
print(f"Correlation between GLP% and AIG count: {correlation1:.2f}")
```

```python
ax = sns.regplot(data=df_corr , x='AIG KI-GR13 PCT', y='glp_pct',scatter_kws={'alpha':0.5},color='green')
plt.title('Correlation between AIG student percent and GLP%')
plt.xlabel('Number of AIG Students')
plt.ylabel('Grade-Level Proficiency (%)')
plt.grid(True)
plt.show()
```
